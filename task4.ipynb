```python
# 1. Import libraries
import pandas as pd
import numpy as np

# 2. Load dataset
df = pd.read_csv('Task 3 and 4_Loan_Data.csv')

# Check columns
print(df.head())
print(df.columns)

# 3. Create 5 rating buckets
# Rating 1 = Best FICO
# Rating 5 = Worst FICO

df['rating'] = pd.qcut(
    df['fico_score'],
    q=5,
    labels=[5, 4, 3, 2, 1]
)

# 4. Display first rows
print(df[['fico_score', 'rating']].head())

# 5. Get bucket boundaries
_, bins = pd.qcut(
    df['fico_score'],
    q=5,
    retbins=True
)

print("Bucket Boundaries:")
print(bins)

# 6. Rating summary
rating_summary = (
    df.groupby('rating')
      .agg(
          min_fico=('fico_score', 'min'),
          max_fico=('fico_score', 'max'),
          avg_fico=('fico_score', 'mean'),
          default_rate=('default', 'mean'),
          customers=('default', 'count')
      )
)

print("\nRating Summary:")
print(rating_summary)

# 7. Create Rating Map
rating_map = (
    df.groupby('rating')
      .agg(
          min_fico=('fico_score', 'min'),
          max_fico=('fico_score', 'max')
      )
)

print("\nRating Map:")
print(rating_map)
```

Borrowers were grouped into five rating buckets using quantile-based discretization of FICO scores. Lower rating numbers correspond to better credit quality. The resulting buckets summarize borrower risk and allow estimation of default rates by rating category.
